````python
def is_decimal_num(test_value):
    """
    Tests if the input is numeric, including numbers with decimals.
    Returns True if value is string type and conforms to above pattern.
    Returns True if value is already float or int type.
    Other than that, returns False.
    """
    if type(test_value) == str:
        allowed = [str(i) for i in range(10)] + ['.']
        if any(c not in allowed for c in test_value):
            return False
        else:
            return True
    elif type(test_value) == int or type(test_value) == float:
        return True
    else:
        return False
````

````python
def standardised_columns(df, excluded_cols):
    """
    Receives a dataframe (df) as input and output a list of column names
    appended with the dataframe name and subsequently
    lowercased, with spaces converted to underscores.
    Does not process column names in excluded_col (list).
    """
    return [(df.name + '_' + col.strip()).replace(' ', '_').lower()
            if col not in excluded_cols
            else col.lower() for col in df.columns]
````

#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code
def sd(sample):
    sample_size = len(sample)
    sample_mean = sum(sample)/sample_size
    return (sum([(x-sample_mean)**2 for x in sample])/sample_size)**0.5
````

To get the rest of the information about states with the lowest and highest participation, SAT total, and ACT composite, we will use a function below so as to minimise repetition. 

````python
def get_high_low_state_info(df, col_list):
    """
    Prints the state(s) with the highest and lowest values 
    of the columns specified in col_list.
    """
    
    def get_vals(series):
        vals_str = ', '.join([v for _, v in series.iteritems()])
        if len(series) == 1:
            return f'is {vals_str}'
        else:
            first_vals, last_vals = vals_str.rsplit(', ', maxsplit=1)[0], vals_str.rsplit(', ', maxsplit=1)[1] 
            return f'are {first_vals} and {last_vals}'
        
    for col in col_list:
        max_states = df[df[col] == df[col].max()]['state']
        min_states = df[df[col] == df[col].min()]['state']
        col = col.replace('_', ' ')
        print(f'The state(s) with the highest {col} {get_vals(max_states)}. \n')
        print(f'The state(s) with the lowest {col} {get_vals(min_states)}. \n')
````
